# Checking data with the Time

## AT

```python
df_at_time = df_combined[['Period Start Time','AT']]
```

```python
df_at_time
```

```python
df_at_time.isnull().sum()
```

```python
sns.histplot(df_at_time)
```

```python
df_at_time['Period Start Time'] = pd.to_datetime(df_at_time['Period Start Time'])
```

```python
df_at_time['Period Start Time']
```

```python
df_at_time['Date'] = df_at_time['Period Start Time'].dt.date
```

```python
df_at_time
```

```python
df_filtered1 = df_at_time[(df_at_time['Period Start Time'].dt.year >= 2019) &
                         (df_at_time['Period Start Time'].dt.year <= 2023)]
```

```python
daily_at = df_filtered1.groupby('Date')['AT'].mean().reset_index()
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(x='Date', y='AT', data=daily_at, linewidth=0.8)
plt.title('Daily Mean AT from 2019 to 2023')
plt.xlabel('Date')
plt.ylabel('Mean AT')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 2019 January

```python
df_filtered = df_at_time[
    (df_at_time['Period Start Time'].dt.year == 2019) &
    (df_at_time['Period Start Time'].dt.month == 1)
]
```

```python
df_filtered['Date'] = df_filtered['Period Start Time'].dt.date
daily_at = df_filtered.groupby('Date')['AT'].mean().reset_index()

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=daily_at, x='Date', y='AT', marker='o')
plt.title('Daily Mean AT - January 2019')
plt.xlabel('Date')
plt.ylabel('Mean AT')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
